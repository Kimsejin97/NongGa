```python
import pandas as pd
```

```python
weather1 = pd.read_csv('./weather1.csv')
weather2 = pd.read_csv('./weather2.csv')
```

```python
weather = pd.concat([weather1, weather2])
```

```python
weather
```

```python
data = weather[weather['지역(군)'].isin(["고창군", "안동시", "고흥군", "의성군", "신안군"])]
```

```python
data
```

```python
data.loc[data["지역(군)"] == "고창군", "지역(군)"] = "Gochang"
data.loc[data["지역(군)"] == "고흥군", "지역(군)"] = "Goheung"
data.loc[data["지역(군)"] == "신안군", "지역(군)"] = "Sinan"
data.loc[data["지역(군)"] == "안동시", "지역(군)"] = "Andong"
data.loc[data["지역(군)"] == "의성군", "지역(군)"] = "Uiseong"
```

```python
data
```

```python
data.to_csv('data.csv', encoding='utf-8', index=False, header=False)
```
